```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
path="/content/drive/MyDrive/Dataset/smartphones.csv"
data = pd.read_csv(path)
```

```python
display(data.head(10))
```

```python
#Scatter plot with RAM against Final price
plt.scatter(data['RAM'], data['Final Price'])
#adding title to the plot
plt. title("Scatter Plot")
#setting the xyand y Labels
plt.xlabel('RAM')
plt.ylabel( 'Final Price')
#adding the Legends
plt.show()
```

```python
#Line chart with RAM against Final price
plt.plot(data['RAM'])
plt.plot(data['Final Price'])
#adding title to the plot
plt. title("Line Chart")
#setting the xyand y Labels
plt.xlabel('RAM')
plt.ylabel( 'Final Price')
#adding the Legends
plt.show()
```

```python
#Bar chart with RAM against Final price
plt.bar(data['RAM'], data['Final Price'])
#adding title to the plot
plt. title("Bar Chart")
#setting the xyand y Labels
plt.xlabel('RAM')
plt.ylabel( 'Final Price')
#adding the Legends
plt.show()
```
